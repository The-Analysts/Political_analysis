```python
ld={} 

for page in range(0,102):
    url = "https://www.latestlaws.com/library/legal-dictionery?page={}".format(page)
    html = urllib.request.urlopen(url)
    soup = BeautifulSoup(html,"html.parser")
    for i in range(2,len(soup.find_all('p'))-1,2):
        ld[soup.find_all('p')[i].text]=soup.find_all('p')[i+1].text
    
```

```python
ld['Aba']=ld.pop('\ufeffAba')
```

```python
import re
for k,v in ld.items():
    ld[k]=re.sub("V|ADJ|N|ABBRV","",v)
```

```python
for k,v in ld.items():
    if v[1]=='1':
        ld[k]=v.split(')')[1][:-2]
    else:
        ld[k]=v.split('(1)')[0]
```
